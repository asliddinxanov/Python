```python
#import Pywhatkit and OpenCv
import pywhatkit as kit
import cv2
```

```python
hand_write = input('Enter your text >> ')
kit.text_to_handwriting(hand_write, save_to = 'xanov-text.png')
image = cv2.imread('xanov-text.png')
cv2.imshow('Python coding', image)
cv2.waitKey(0)
cv2.destroyAllWindows()
```
